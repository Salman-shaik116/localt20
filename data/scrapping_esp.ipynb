```python
import pandas as pd
import requests
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
teams_table=pd.DataFrame()
for i in range(1,5):
    url = f"https://stats.espncricinfo.com/ci/engine/stats/index.html?class=3;filter=advanced;orderby=runs;page={i};size=200;spanmax1=25+Oct+2024;spanmax2=25+Oct+2024;spanmin1=25+Oct+2014;spanmin2=25+Oct+2014;spanval1=span;spanval2=span;template=results;type=team;view=season"
    response = requests.get(url, headers=headers)
    response.raise_for_status()  # Raise an exception for HTTP errors
    # Read HTML with pandas
    tables = pd.read_html(response.text, flavor='bs4')  # or use flavor='html5lib'
    stats_table = tables[2]
    teams_table=pd.concat([stats_table, teams_table], axis=0, ignore_index=True)
```

```python
teams_table.sort_values(by='Team', ascending=False, inplace=True)
```

```python
teams_table.drop(columns=['Unnamed: 14'], axis=1,inplace=True)
teams_table
```

```python
teams_table.to_csv('t20_team_stats.csv')
```
